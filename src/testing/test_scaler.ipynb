```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PowerTransformer, QuantileTransformer, RobustScaler
import random
```

```python
size = 1000
```

```python
x1 = [random.uniform(-1000, 10000) for _ in range(size)]
x2 = [random.uniform(-1, 1) for _ in range(size)]
x3 = [random.uniform(-1, 1) for _ in range(size)]
x4 = [random.uniform(-1, 1) for _ in range(size)]
x5 = [random.uniform(-1, 1) for _ in range(size)]
x6 = [random.uniform(-1, 1) for _ in range(size)]
x7 = [random.uniform(-1, 1) for _ in range(size)]
x8 = [random.uniform(-1, 1) for _ in range(size)]
x9 = [random.uniform(-1, 1) for _ in range(size)]
x10 = [random.uniform(-1, 1) for _ in range(size)]
```

```python
data = np.array([x1,x2,x3,x4, x5, x6, x7, x8, x9, x10]*100)
data.shape
```

```python
# scaler = MinMaxScaler()
# scaler = StandardScaler()
# scaler = PowerTransformer()
scaler = QuantileTransformer()
# scaler = RobustScaler()
scaler.fit(data)
```

```python
scaled_data = scaler.transform(data)
```

```python
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
pred1 = np.array([[random.uniform(-1, 1) for _ in range(size)]])
```

```python
res1 = pca.transform(pred1)
```
