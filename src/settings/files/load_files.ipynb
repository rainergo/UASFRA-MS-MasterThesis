```python
import pandas as pd
```

```python
df_de = pd.read_csv('1000_german_most_common_words', sep='\t', index_col=0, header=0).sort_values(by='German', ascending=True)
df_de
```

```python
df_en = pd.read_csv('1000_english_most_common_words', header=0, encoding='utf-8').sort_values(by='WORD', ascending=True)
df_en
```

```python
de  = df_de.German.tolist()
en = df_en.WORD.tolist()
```

```python
df = pd.DataFrame({'en': en, 'de': de})
df['en'] = df.en.str.lower()
df['de'] = df.de.str.lower()
```

```python
df
```

```python
df.to_parquet('1000_most_common_words.parquet')
```
